```python
class NodoBinario:
    def __init__(self, value, left = None, right = None):
        self.data = value
        self.left = left
        self.right = right

class NodoTernario:
    def __init__(self, value, left = None, center = None, right = None ):
        self.data = value
        self.left = left
        self.center = center
        self.right = right
        
```

```python
arbol1 = NodoBinario("R", NodoBinario("C"), NodoBinario ("H"))
print(arbol1.right.data)

arbol2 = NodoTernario(4, NodoTernario(3, NodoTernario(5, left = NodoTernario(2)), NodoTernario(5),NodoTernario(2)), None ,NodoTernario(5))
print(arbol2.left.right.data)
print(arbol2.data)
print(arbol2.left.center.data)
print(arbol2.right.left)
```

```python
print(arbol2.left.center.data)
print(arbol2.left.right.data)
print(arbol2.left.left.left.data)
print("---- sin usar la palabra left en el print ----")
ejemplo = arbol2.left.left.left.data
print(ejemplo)
print("---- otra forma? ----")
aux = arbol2
while(aux.left != None):
    aux = aux.left
print(aux.data)
```
